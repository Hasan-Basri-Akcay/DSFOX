```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_diabetes, load_iris, load_boston
```

# Classification

```python
clf_data = load_iris()
df_clf = pd.DataFrame(data=clf_data.data, columns=clf_data.feature_names)
df_clf['random_cat_col'] = np.random.choice(['A', 'B', 'C'], size=len(df_clf))
df_clf['target'] = clf_data['target']
df_clf.head()
```

# Regression

```python
reg_data = load_diabetes()
df_reg = pd.DataFrame(data=reg_data.data, columns=reg_data.feature_names)
df_reg['target'] = reg_data['target']
df_reg['sex'] = df_reg['sex'].astype(str)
df_reg['sex'].replace({'0.0506801187398187':'M', '-0.044641636506989':'F'}, inplace=True)
df_reg.head()
```
